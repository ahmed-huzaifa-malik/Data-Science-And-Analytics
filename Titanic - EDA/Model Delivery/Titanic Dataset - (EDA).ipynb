```python
# Import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
titanic_data = pd.read_csv('/content/titanic.csv')
```

```python
# Show the first few rows of the dataset
titanic_data.head()
```

```python
# Show the last few rows of the dataset
titanic_data.tail()
```

```python
# Show unqiue columns of the dataset
titanic_data.columns.unique()
```

```python
# Show some basic statistics for the dataset
titanic_data.describe()
```

```python
# Show basic info about the dataset
titanic_data.info()
```

```python
# Show the total rows and columns of the dataset
titanic_data.shape
```

```python
# Show missing values of columns of the dataset
titanic_data.isnull()
```

```python
# Sum up the missing values of each column
titanic_data.isnull().sum()
```

```python
# Now take the Fare column into consideration
mean = titanic_data['Fare'].mean()
median = titanic_data['Fare'].median()
mode = titanic_data['Fare'].mode()

print(f"The mean of Fare is: {mean:.2f}")
print(f"The median of Fare is: {median:.2f}")
print(f"The mode of Fare is: {mode:}")
```

```python
# As the value of median is less sensitive to outliers so we fill the missing values with this
titanic_data['Fare'] = titanic_data['Fare'].fillna(median)
```

```python
# Check for duplicated values if any
titanic_data.duplicated()
```

```python
# Sum up the dulicated values if any
titanic_data.duplicated().sum()
```

```python
# Now take Age column into consideration
mean = titanic_data['Age'].mean()
median = titanic_data['Age'].median()
mode = titanic_data['Age'].mode()[0]

print(f"The mean of Age is: {mean:.2f}")
print(f"The median of Age is: {median:.2f}")
print(f"The mode of Age is: {mode:}")
```

```python
# Now we will take mode as its value is less sensitive to outliers and fill the missing values with this
titanic_data['Age'] = titanic_data['Age'].fillna(mode)
```

```python
# Now check again for missing values if any
titanic_data.isnull()
```

```python
# Sum up to check missing values if any
titanic_data.isnull().sum()
```

```python
# Drop the Cabin column as it is of no concern to our use
titanic_data.drop('Cabin', axis=1, inplace=True)
```

```python
# Check again for verification
titanic_data.isnull().sum()
```

```python
# Show the Bar Chart of how many passengers survived and how many did not
sns.countplot(x='Survived', data=titanic_data)
plt.title('Count of Passengers Survived and Not Survived')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Show the Bar Chart of how many males and females were present
sns.countplot(x='Sex', data=titanic_data)
plt.title('Count of Males and Females')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
# Show the Bar chart for 'Pclass'
sns.countplot(x='Pclass', data=titanic_data)
plt.title('Count of Passengers by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
# Show the Histogram for 'Age'
plt.hist(titanic_data['Age'], bins=20)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Show the Count Plot of Survivals by Passenger Class
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data = titanic_data)
plt.title('Survivals by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
# Show the Count Plot of Survivals By Gender
plt.figure(figsize=(8, 6))
sns.countplot(data = titanic_data, x='Sex', hue='Survived', palette='pastel')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

```python
# Show the Histogram of Survivals By Age
sns.histplot(data = titanic_data, x='Age', hue='Survived', kde=True, multiple='stack')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.title('Age Distribution by Survival')
plt.show()
```

```python
# Show Pair Plot for detailed analysis
sns.pairplot(titanic_data[['Age', 'Sex', 'Fare', 'SibSp', 'Parch', 'Survived']])
plt.show()
```

```python
sns.pairplot(titanic_data[['Age', 'Sex', 'Fare', 'SibSp', 'Parch', 'Survived']], hue = 'Survived')
plt.figure(figsize=(10, 10))
plt.show()
```

```python
# Box Plot for Fare for Outlier Detection
sns.boxplot(x=titanic_data['Fare'])
plt.title('Box Plot of Fare')
plt.show()
```

```python
# Covert values of Sex to numeric
titanic_data['Sex'] = titanic_data['Sex'].map({'male': 0, 'female': 1})
```

```python
# Select only numeric columns
numeric_data = titanic_data.select_dtypes(include=['number'])
# Compute correlation matrix
corr = numeric_data.corr()
plt.figure(figsize=(10, 9))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
